```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
mean= (-3,0)
cov=[[1.0,0.8],[0.8,1.0]]
size=(2,500)
x,y=np.random.multivariate_normal(mean,cov,size).T
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
# plt.axis('equal')
plt.scatter(x,y)
plt.show()
```

```python
  # histgram
fig, ax = plt.subplots()
ax.set_xlabel('x1')
ax.set_ylabel('frequency')
ax.hist(x)
plt.xlim(-5.5,-1)
plt.show()
```

```python
fig, ay = plt.subplots()
ay.set_xlabel('y1')
ay.set_ylabel('frequency')
ay.hist(y)
plt.xlim(-2.5,2.5)
plt.show()
```

```python
mean2= (0,3)
x1,y1=np.random.multivariate_normal(mean2,cov,size).T
# plt.axis('equal')
plt.scatter(x1,y1, color="orange", label="rotated")
plt.scatter(x,y, label="original")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot with Two Colors')
plt.legend()
plt.show()
```

```python
z=np.vstack((y1,x))
np.shape(z)
```
