#Problem 5

```python
def function1(x):
  return (1/2)*x+1

def function2(x):
  return x**2

def function3(x):
  return 2*(x**2)+2**x

def function4(x):
  return np.sin(x**(1/2))

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):

  if(function==4):
    x=np.arange(0, x_range[1], x_range[2])
    y=np.arange(0, x_range[1], x_range[2])
  else:
    x=np.arange(x_range[0], x_range[1], x_range[2])

  if(function==1):
    y=function1(x)
  elif(function==2):
    y=function2(x)
  elif(function==3):
    y=function3(x)
  elif(function==4):
    y=function4(x)

  colab=np.array([x,y])
  colab=np.reshape(colab,len(x)*2, order='F')
  newArray=colab.reshape((len(x),2))

  first=0
  second=len(x)-1
  x_side=newArray[second][0]-newArray[first][0]
  y_side=newArray[second][1]-newArray[first][1]
  gradient_slope=y_side/x_side

  return newArray,gradient_slope


def plotShow(x, y):
  plt.plot(x, y)
  plt.title("Slope")
  plt.xlabel("X")
  plt.ylabel("Y Slope")
  plt.show()
```

```python
news,grad=compute_gradient(1)
print(news)
print(grad)
plotShow(news[0:,0],news[0:,1])
```

```python
news,grad=compute_gradient(2)
print(news)
print(grad)
plotShow(news[0:,0],news[0:,1])
```

```python
news,grad=compute_gradient(3)
print(news)
print(grad)
plotShow(news[0:,0],news[0:,1])
```

```python
news,grad=compute_gradient(4)
print(news)
print(grad)
plotShow(news[0:,0],news[0:,1])
```

#Problem 6(Finding minimum value and taking it of the array)
