#Problem 2, 5

```python
def board(row,column):
  n_squares = row*column
  small_board_list = [1]

  # to count the average
  average_list=[]

  # to divide and sum each half of the chessboard
  half=n_squares/2
  half_sum=[0,0]
  sum_counter=0

  # for to create the values in that will create the matrix
  for i in range(n_squares - 1):
      small_board_list.append(2*small_board_list[-1])

      # problem 5 getting each half sum of the upper and lower part of the matrix
      if sum_counter<half:
        half_sum[0]=half_sum[0]+small_board_list[i]
      else:
        half_sum[1]=half_sum[1]+small_board_list[i]
      sum_counter=sum_counter+1

  # adding th last value
  half_sum[1]=half_sum[1]+small_board_list[n_squares-1]

  # converting half_sum and smallboard to np.array
  small_board_ndarray = np.array(small_board_list)
  half_sum_ndarray = np.array(half_sum)

  # reshaping them to numPy array bidimensional
  c1=small_board_ndarray.reshape(row,column)

  # getting the sum of the wheats of all the grains in all the chessboard spaces
  sum=0
  for i in range(column):
    for j in range(row):
      sum=sum+c1[j][i]
    average_list.append(sum/row)
  sum=np.sum(c1)
  average_list=np.array(average_list)
  return c1,sum,average_list,half_sum_ndarray

chessboard,sum,average_list,half_sum_ndarray=board(2,2)
print(chessboard)
print(sum)
print(average_list)
print(half_sum_ndarray)
print("The first half of value {} fits in the second half of value {} a total of {} times".format(half_sum_ndarray[0],half_sum_ndarray[1],half_sum_ndarray[1]/half_sum_ndarray[0]))
```

#Problem 6

```python
def broadCastCappendMatrix(columns,rows):
  n_squares=columns*rows
  board_ndarray = np.array([1])
  for _ in range(n_squares - 1):
      board_ndarray = np.append(board_ndarray, 2*board_ndarray[-1])
  indices_of_squares = np.arange(n_squares).astype(np.uint64)
  board_ndarray = 2**indices_of_squares
  return board_ndarray
```

```python
%%timeit
values=broadCastCappendMatrix(8,8)
print(values)
```
