```python
import matplotlib.pyplot as plt
# Chestnut bun volume in cubic centimeters
CBVolume=283

# milky way volume in cubic AU
SSVolume=5.24*(10**14)
print(SSVolume)
```

```python
# function that converts from AU to centimeters
def AU_centimeter_converter(value_AU):
  return value_AU*(3.35*(10**39))

SSCentimeterVolume=AU_centimeter_converter(SSVolume)
print("Solar System value in cubic AU:{}".format(SSVolume))
print("Solar System value in cubic centimeters:{}".format(SSCentimeterVolume))
```

```python
# function that counts how many iterations it takes to fill up an area
def filling_space_count(fill_area,fill_item,time_unit):
  area_volume=fill_item
  iteration=0
  minutesList=[0]
  area_volume_instance=[fill_item]
  while fill_area>area_volume:
    area_volume+=fill_item*(2**iteration)
    iteration+=1
    minutesList.append(iteration*time_unit)
    area_volume_instance.append(area_volume)
  minutes=minutesList[-1]
  return iteration, minutes, minutesList, area_volume_instance

def graph_plot(minutesList,area_volume_instance):
  plt.title("Area being filled")
  plt.xlabel("Minutes")
  plt.ylabel("Volume")
  plt.plot(minutesList,area_volume_instance )  # 「リスト名」のところにリストの変数名を入れる
  plt.show()

[iteration, minutes, minutesList, area_volume_instance]=filling_space_count(2000,CBVolume,5)
print("In the following example, it's filling up a space of 2000 cubic cm with 283 cubic cm Chestnut buns, which took {} iterations, that amounts to {} minutes".format(iteration,minutes))

graph_plot(minutesList,area_volume_instance)
```

```python
# Now to fill up the solar system with chestnut buns
[iteration, minutes, minutesList, area_volume_instance]=filling_space_count(SSCentimeterVolume,CBVolume,5)
print("In the following example, it's filling up a space of Solar System {} cubic cm with {} cubic cm Chestnut buns, which took {} iterations, that amounts to {} minutes".format(SSCentimeterVolume,CBVolume,iteration,minutes))
graph_plot(minutesList,area_volume_instance)
```

```python
# Football volume filling up the Tokyo Dome
TDVolume=1.24*(10**12)
FBVolume=5792
[iteration, minutes, minutesList, area_volume_instance]=filling_space_count(TDVolume,FBVolume,5)
print("In the following example, it's filling up a space of Tokyo Dome {} cubic cm with {} cubic cm Football, which took {} iterations, that amounts to {} minutes".format(TDVolume,FBVolume,iteration,minutes))
graph_plot(minutesList,area_volume_instance)
```
