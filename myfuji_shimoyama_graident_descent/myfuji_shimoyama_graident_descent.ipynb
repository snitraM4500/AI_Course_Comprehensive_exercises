```python
import numpy as np

csv_path = "mtfuji_data.csv" # ファイル名（パス）を指定する

np.set_printoptions(suppress=True) # 指数表示の禁止を設定する
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
```

```python
number_of_the_point=np.copy(fuji[0:,0])
latitude=np.copy(fuji[0:,1])
longitude=np.copy(fuji[0:,2])
altitude=np.copy(fuji[0:,3])
distance_from_0=np.copy(fuji[0:,4])
```

```python
import matplotlib.pyplot as plt

plt.plot(number_of_the_point, altitude)
plt.title("Mt. Fuji")
plt.xlabel("position")
plt.ylabel("elevation[m]")
plt.show()
```

```python
max=len(number_of_the_point)
gradient_slope=[]
for i in range(max):
  if i<max-1:
    first=i
    second=i+1
    x_side=number_of_the_point[second]-number_of_the_point[first]
    y_side=altitude[second]-altitude[first]
    gradient_slope.append(y_side/x_side)
print(gradient_slope)
```

```python
def calculate_gradient(current_point):
    if current_point < 0 or current_point >= len(fuji) - 1:
        return None

    current_elevation = fuji[current_point, 3]
    next_elevation = fuji[current_point + 1, 3]

    current_point_number = fuji[current_point, 0]
    next_point_number = fuji[current_point + 1, 0]

    gradient = (next_elevation - current_elevation) / (next_point_number - current_point_number)
    return gradient
```

```python
def calculate_next_point(current_point, alpha=0.2):
    gradient = calculate_gradient(current_point)

    if gradient is None:
        return None

    next_point = current_point - alpha * gradient
    next_point = round(next_point)

    if next_point < 0 or next_point >= len(fuji):
        return None

    return int(next_point)
```

```python
current_point = 136
num_iterations = 220
def iterating_process(current_point,num_iterations):
  points_moved = [current_point]
  elevations = [fuji[current_point, 3]]
  gradients = [calculate_gradient(current_point)]
  for i in range(num_iterations):
      current_point = calculate_next_point(current_point)
      if current_point is None:
          break
      points_moved.append(current_point)
      elevations.append(fuji[current_point, 3])
      gradients.append(calculate_gradient(current_point))
  return points_moved

points_moved=iterating_process(current_point,num_iterations)
plt.figure(figsize=(10, 6))
plt.plot(fuji[:, 4], fuji[:, 3], label="Mt. Fuji Profile")
plt.plot(fuji[points_moved, 4], fuji[points_moved, 3], 'ro', label="Moved Points")
plt.xlabel("Distance from Point 0 (m)")
plt.ylabel("Elevation (m)")
plt.title("Descent Process on Mt. Fuji Profile")
plt.legend()
plt.grid()
plt.show()
```
