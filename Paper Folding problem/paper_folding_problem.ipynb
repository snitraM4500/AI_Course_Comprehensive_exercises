# Problem 3

```python
NUMBER_OF_FOLD=43
interaction=1
for i in range(NUMBER_OF_FOLD+1):
    folded_thickness = THICKNESS * interaction
    interaction *= 2
    if(folded_thickness >= 1000):
        print("Thickness of {} fold： {:.2f} kilometers".format(i,folded_thickness/1000))
    else:
        print("Thickness of {} fold： {} meters".format(i,folded_thickness))
```

# Problem 4

```python
%%timeit
import time
THICKNESS = 0.00008
NUMBER_OF_FOLD=43
print("Timing operations of {} folds:".format(NUMBER_OF_FOLD))
print()

print("-------------------Timing with exponetiation-------------------")
start = time.time()
folded_thickness = THICKNESS * 2**NUMBER_OF_FOLD
print("Thickness： {} meters".format(folded_thickness))
elapsed_time = time.time() - start
print("Total time with exponatiation: {}[s]".format(elapsed_time))
print()

print("-------------------Timing with for loop-------------------")
start = time.time()
interaction=1
for i in range(NUMBER_OF_FOLD+1):
    folded_thickness = THICKNESS * interaction
    interaction *= 2
elapsed_time = time.time() - start
print("Thickness： {} meters".format(folded_thickness))
print("Total time with for loop : {}[s]".format(elapsed_time))


NUMBER_OF_FOLD=500
print()
print("Timing operations of {} folds:".format(NUMBER_OF_FOLD))
print()

print("-------------------Timing with exponetiation-------------------")
start = time.time()
folded_thickness = THICKNESS * 2**NUMBER_OF_FOLD
print("Thickness： {} meters".format(folded_thickness))
elapsed_time = time.time() - start
print("Total time with exponatiation: {}[s]".format(elapsed_time))
print()

print("-------------------Timing with for loop-------------------")
start = time.time()
interaction=1
for i in range(NUMBER_OF_FOLD+1):
    folded_thickness = THICKNESS * interaction
    interaction *= 2
elapsed_time = time.time() - start
print("Thickness： {} meters".format(folded_thickness))
print("Total time with for loop : {}[s]".format(elapsed_time))
```

# Problem 6 & 7

```python
import matplotlib.pyplot as plt

PAPER_THICKNESS=0.00008
number_of_fold=44
values_per_fold=[]
iterations=[]
for i in range(number_of_fold):
    iterations.append(i)
    values_per_fold.append(PAPER_THICKNESS*(2**i))
print(len(values_per_fold))

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(values_per_fold )  # 「リスト名」のところにリストの変数名を入れる
plt.show()

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(iterations,values_per_fold, color="red")  # 「リスト名」のところにリストの変数名を入れる
plt.show()

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=20)
plt.plot(iterations,values_per_fold )  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```
